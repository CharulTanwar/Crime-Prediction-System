#  SimpleImputer to replace missing values (NaN)

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 2:31])  
x[:, 2:31] = imputer.transform(x[:, 2:31])  

print(x)
```

# Performing One hot encoding in state column

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from seaborn import load_dataset
import pandas as pd

df = x

transformer = make_column_transformer(
    (OneHotEncoder(), [0]),
    remainder='passthrough')

transformed = transformer.fit_transform(df)
x = transformed
print(x)
```

# Training the K-Means model on the dataset

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42, n_init = 10)
y_kmeans = kmeans.fit_predict(A)
print(y_kmeans)
```

#  Initializes a sequential neural network model using TensorFlow's Keras API.

```python
import numpy as np
import pandas as pd
import tensorflow as tf
ann = tf.keras.models.Sequential()
```

# Adding hidden layer configuration

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

# Compiling Neural Network

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Training Neural Network

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
x_train = np.asarray(x_train).astype('float32')
y_train = np.asarray(y_train).astype('float32')
ann.fit(x_train, y_train, batch_size = 32, epochs = 100)
```

# Neural Network Prediction Evaluation

```python
x_test = np.asarray(x_test).astype('float32')
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Accuracy score

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



states = dataset['STATE/UT'].unique()

for state in states:
    print(state, " : \n")
    data = dataset[dataset['STATE/UT'] == state]
    grouped = data.groupby('YEAR')['TOTAL IPC CRIMES'].sum()
    year = grouped.index
    arr = grouped.values
    plt.figure()
    plt.plot(year, arr)
    plt.title(f"Total IPC Crimes in {state} over the years")
    plt.xlabel('Year')
    plt.ylabel('Total IPC Crimes')
    plt.show()
    print("\n")
```
